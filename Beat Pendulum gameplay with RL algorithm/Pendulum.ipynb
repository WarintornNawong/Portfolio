```python
import gym
from stable_baselines3 import PPO,A2C,DDPG,SAC,TD3
from stable_baselines3.common.noise import NormalActionNoise
from stable_baselines3.common.evaluation import evaluate_policy

import matplotlib.pyplot as plt
import numpy as np
```

```python
env = gym.make('Pendulum-v1', g=9.81)
```

```python
def average_reward(reward):   
    start = np.arange(0,991)
    mean = []
    for i in start:
        end = i + 10
        mean.append(np.mean(reward[0][i:end]))
    return mean
```

## Part I. Model Construction 
### I. Proximal Policy Optimization (PPO)

```python
pre_ppo = PPO(policy = "MlpPolicy",env =  env, verbose=1)
```

```python
pre_ppo_reward = evaluate_policy(pre_ppo, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(pre_ppo_reward[0])
```

```python
post_ppo = PPO(policy = "MlpPolicy",env =  env, verbose=1)
post_ppo.learn(total_timesteps=25000)
```

```python
post_ppo_reward = evaluate_policy(post_ppo, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(post_ppo_reward[0])
```

```python
pre_PPO_mean = average_reward(pre_ppo_reward)
post_PPO_mean = average_reward(post_ppo_reward)
```

### II.Advantage Actor-Critic (A2C)

```python
pre_a2c = A2C(policy = "MlpPolicy",env =  env, verbose=1)
```

```python
pre_a2c_reward = evaluate_policy(pre_a2c, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(pre_a2c_reward[0])
```

```python
post_a2c = A2C(policy = "MlpPolicy",env =  env, verbose=1)
post_a2c.learn(total_timesteps=25000)
```

```python
post_a2c_reward = evaluate_policy(post_a2c, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(post_a2c_reward[0])
```

```python
pre_A2C_mean = average_reward(pre_a2c_reward)
post_A2C_mean = average_reward(post_a2c_reward)
```

### III. Deep Deterministic Policy Gradient (DDPG)

```python
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))
pre_DDPG = DDPG("MlpPolicy", env, action_noise=action_noise, verbose=1)
```

```python
pre_ddpg_reward = evaluate_policy(pre_DDPG, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(pre_ddpg_reward[0])
```

```python
post_DDPG = DDPG("MlpPolicy", env, action_noise=action_noise, verbose=1)
post_DDPG.learn(total_timesteps=25000, log_interval=10)
```

```python
post_ddpg_reward = evaluate_policy(post_DDPG, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(post_ddpg_reward[0])
```

```python
pre_DDPG_mean = average_reward(pre_ddpg_reward)
post_DDPG_mean = average_reward(post_ddpg_reward)
```

### IV. Soft Actor-Critic (SAC)

```python
pre_SAC = SAC("MlpPolicy", env, verbose=1)
```

```python
pre_SAC_reward = evaluate_policy(pre_SAC, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(pre_SAC_reward[0])
```

```python
post_SAC = SAC("MlpPolicy", env, verbose=1)
post_SAC.learn(total_timesteps=25000, log_interval=4)
```

```python
post_SAC_reward = evaluate_policy(post_SAC, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(post_SAC_reward[0])
```

```python
pre_SAC_mean = average_reward(pre_SAC_reward)
post_SAC_mean = average_reward(post_SAC_reward)
```

###  V.Twin Delayed Deep Deterministic Policy Gradient (TD3)

```python
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))

pre_TD3 = TD3("MlpPolicy", env, action_noise=action_noise, verbose=1)
```

```python
pre_TD3_reward = evaluate_policy(pre_TD3, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(pre_TD3_reward[0])
```

```python
post_TD3 = TD3("MlpPolicy", env, action_noise=action_noise, verbose=1)
post_TD3.learn(total_timesteps=25000, log_interval=10)
```

```python
post_TD3_reward = evaluate_policy(post_TD3, env, n_eval_episodes=1000,return_episode_rewards=True)
```

```python
plt.plot(post_TD3_reward[0])
```

```python
pre_td3_mean = average_reward(pre_TD3_reward)
post_td3_mean = average_reward(post_TD3_reward)
```
